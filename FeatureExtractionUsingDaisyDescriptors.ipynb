```python
import skimage
from skimage import color 
from skimage.feature import daisy
from skimage import data
from skimage.transform import resize

import matplotlib.pyplot as plt
```

```python
 horse_gray = color.rgb2gray(skimage.io.imread("./horse.jpg"))
```

```python
horse = resize(horse_gray, output_shape=(300, 450), mode='reflect', anti_aliasing=False)

plt.figure(figsize=(8, 8))
plt.imshow(horse, cmap='gray')
```

```python
descs_array, descs_image = daisy(horse, step=150, radius=30, visualize=True)
```

```python
descs_num = descs_array.shape[0] * descs_array.shape[1]

descs_num
```

```python
plt.figure(figsize=(8, 8))

plt.title("Daisy Descriptors")
plt.imshow(descs_image)
```
